```python
loans_itr = pd.read_csv('loans_2007.csv', chunksize=3000)
list_of_uniq = []

for chunk in loans_itr:
    values = {}
    string_cols = chunk.select_dtypes(include=['object'])
    for col in string_cols:
        length = chunk[col].size
        uniq = chunk[col].unique().size
        values[col] = [uniq/length]
    part_uniq = pd.DataFrame.from_dict(values)
    list_of_uniq.append(part_uniq)
total_uniq = pd.concat(list_of_uniq)

        
        
```

```python
total_uniq.mean().sort_values(ascending=True)
```

```python
loans_itr = pd.read_csv('loans_2007.csv', chunksize=3000)
list_of_na = []

for chunk in loans_itr:
    values = {}
    float_cols = chunk.select_dtypes(include=['number'])
    for col in float_cols:
        #length = chunk[col].size
        #len(chunk[col]) - chunk[col].count()   
        values[col]= [chunk[col].isnull().sum()]
    part_na = pd.DataFrame.from_dict(values)
    list_of_na.append(part_na)
total_na = pd.concat(list_of_na)
```

```python
#Total NA values
total_na.sum().sort_values(ascending=False)
```
